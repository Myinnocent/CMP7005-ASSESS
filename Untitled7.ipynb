**LOOKING AT MISSING VALUE**

```python
# Missing values
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()

        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)

        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        print(mis_val_table)

        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})

        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns.sort_values(
        '% of Total Values', ascending=False)

        # Return the dataframe with missing information
        return mis_val_table_ren_columns

missing_values= missing_values_table(df)
missing_values.style.background_gradient(cmap='Blues')
```

**visualizing each feature to see the distribution using a subplot**

```python
# Visualize all features using subplots
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(15, 20))
axes = axes.flatten()

# List of features to plot
features = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

# Plot each feature
for i, feature in enumerate(features):
    sns.histplot(df[feature].dropna(), bins=30, ax=axes[i], kde=True)
    axes[i].set_title(feature)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select numerical columns for visualization
numerical_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

# Create subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
axes = axes.flatten()

for i, column in enumerate(numerical_columns):
    ax = axes[i]
    df[column].hist(bins=50, ax=ax, color='blue', alpha=0.7)
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Remove empty subplots
for j in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Imputing missing values based on the disribution of the features**

```python
# Impute missing values with median for skewed distributions
for column in ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'RAIN']:
    df[column] = df[column].fillna(df[column].median())

# Impute missing values with mean for unskewed distributions
for column in ['TEMP', 'PRES', 'DEWP', 'WSPM']:
    df[column] = df[column].fillna(df[column].mean())
df
```

**Monthwise Plot**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df has 'station' and pollutant columns (e.g., 'PM2.5', 'PM10', etc.)

# No need to convert 'station' to datetime as it's already a station identifier
# df['station'] = pd.to_datetime(df['station'], errors='coerce')  # Remove this line

# No need to extract station, it's already in the column
# df['station'] = df['station'].dt.station  # Remove this line

# 3. Group by station and calculate mean for pollutants
pollutant_columns = ['PM2.5', 'PM10', 'NO2', 'CO', 'SO2', 'O3'] # Replace with your actual pollutant column names
df_monthly = df.groupby('station')[pollutant_columns].mean().reset_index()

# 4. Create the plot using Matplotlib
plt.figure(figsize=(12, 6))
for pollutant in pollutant_columns:
    # Use df_monthly['station'] for the x-axis instead of 'Month'
    plt.plot(df_monthly['station'], df_monthly[pollutant], label=pollutant, marker='o')

plt.title("Month-wise Pollutant Levels")
plt.xlabel("station") # Changed to 'station'
plt.ylabel("Pollutant Concentration (µg/m³)")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Assuming df and pollutants are already defined
# Group by Year and Month to calculate the monthly average for each pollutant
monthly_avg = df.groupby(['year','month', 'day'])[pollutants].mean().reset_index() #Added 'year' to groupby

# Create a Date column from Year and Month
monthly_avg["Date"] = pd.to_datetime(monthly_avg[['year','month','day']].assign(DAY=1)) #changed column name from year to Date

# Create subplots for each pollutant
fig = make_subplots(rows=len(pollutants), cols=1, subplot_titles=[f'{pollutant} Monthly Average Concentration Over Time' for pollutant in pollutants])

# Add traces for each pollutant
for i, pollutant in enumerate(pollutants):
    fig.add_trace(
        go.Scatter(
            x=monthly_avg['Date'],
            y=monthly_avg[pollutant],
            mode='lines+markers',
            name=pollutant,
            line=dict(color='blue', width=2),
            marker=dict(size=8),
            opacity=0.7
        ),
        row=i+1, col=1
    )

# Update layout
fig.update_layout(
    title_text='Monthly Average Concentrations of Pollutants Over Time',
    title_font_size=24,
    showlegend=False,
    height=300 * len(pollutants),  # Adjust height based on the number of pollutants
    width=1000
)

# Update y-axis labels
for i, pollutant in enumerate(pollutants):
    fig.update_yaxes(title_text=f'{pollutant} (ug/m3)', row=i+1, col=1)

# Update x-axis labels
fig.update_xaxes(title_text='Date', row=len(pollutants), col=1)

# Rotate x-axis labels
fig.update_xaxes(tickangle=45)

# Show the plot
fig.show()
```
